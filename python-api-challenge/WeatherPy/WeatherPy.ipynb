```python
import matplotlib.pyplot as plt
import numpy as np 
import requests
import pandas as pd
from api_keys import weather_api_key
from citipy import citipy
from scipy.stats import linregress
# output csv data
output_csv_path="data/cities.csv"
# define a random len
random_len=2000

# define range Lat and Lng
lat_range = (-90, 90)
lng_range = (-180, 180)
cities=[]
```

## Generate Cities data

```python
# rundom Lats
lats=np.random.uniform(lat_range[0],lat_range[1],random_len)
lngs=np.random.uniform(lng_range[0],lng_range[1],random_len)
all_alns = zip(lats, lngs)

for lat_lng in all_alns:
    city = citipy.nearest_city(lat_lng[0],lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
len(cities)
```

### using API Inovker
- 1. get all citites data

```python
print("start invoke data from api")
#define a funcation to invoke weather api
def weather_invoke(city):
    url = "http://api.openweathermap.org/data/2.5/weather?"
    query_url = f"{url}appid={weather_api_key}&q={city}&units=imperial"
    response = requests.get(query_url).json()
    return response


city_weather=[]
# start invoke data
index=0
for city in cities:
    index+=1
    print(f"get {index} city :{city} data start")
    data = weather_invoke(city)
    data
    if data['cod']==200:
        city_weather.append({"City": city,
                                     "Lat" : data["coord"]["lat"],
                                     "Lng" : data["coord"]["lon"],
                                     "Temp" : data["main"]["temp"],
                                     "Max Temp" : data["main"]["temp_max"],
                                     "Humidity" : data["main"]["humidity"],
                                     "Cloudiness" : data["clouds"]["all"],
                                     "Wind Speed" : data["wind"]["speed"],
                                     "Country" : data["sys"]["country"],
                                     "Date" : data["dt"]
                            })
    else:
        print('city not found ! skipping!!!!')
print("--------------------")
print("get data is complete")
print("--------------------")
```

### cover  data to csv and using pd print a data

```python
wdf=pd.DataFrame(city_weather)
wdf.count()
wdf.describe()
```

```python
wdf.to_csv(output_csv_path,index=True,header=True)
```

## separating data to Northern and Southern

```python
north_df = wdf.loc[wdf["Lat"] >= 0]
north_df
south_df = wdf.loc[wdf["Lat"] < 0]
south_df
```

### - Northern Hemisphere - Temperature (F) vs. Latitude

```python
 
x_values = north_df["Lat"]
y_values = north_df["Temp"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(2,280),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.title("Southern  Hemisphere - Temperature (F) vs. Latitude (2020/10/24)")
print(f"The r-squared is: {rvalue**2}")
plt.savefig('./data/Northern_Temp_Lat.png')
plt.show()
```
